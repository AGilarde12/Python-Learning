# Plotting Simple Moving Averages

This is a quick analysis of MSFT and AAPL's Simple Moving Average on a 20-day and 50-day frequency

```python
import numpy as np
import pandas as pd
import pandas_datareader as wb
import matplotlib
```

```python
tickers = [
    'MSFT','AAPL'
];

mydata = pd.DataFrame()
for t in tickers:
    mydata[t] = wb.DataReader(t, data_source='yahoo',
    start = '2015-1-1')['Adj Close'];
    
#This prints the first row of the entire data
print(mydata.iloc[0])
```

```python
aapl_sma20 = mydata['AAPL'].rolling(20).mean()
aapl_sma50 = mydata['AAPL'].rolling(50).mean()
msft_sma20 = mydata['MSFT'].rolling(20).mean()
msft_sma50 = mydata['MSFT'].rolling(50).mean()
```

```python
pd.DataFrame({   
    'AAPL': mydata['AAPL'],
    'AAPL_SMA20': aapl_sma20,
    'AAPL_SMA50': aapl_sma50,
    'MSFT': mydata['MSFT'],
    'MSFT_SMA20' : msft_sma20,
    'MSFT_SMA50' : msft_sma50
}).plot(
    title='AAPL & MSFT Close Price / SMA Crossover', figsize = (15,6)
);
```
